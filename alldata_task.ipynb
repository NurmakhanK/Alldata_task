```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url ='https://www.goszakup.gov.kz/ru/registry/rqc'
session = requests.Session()
code_site = session.get(url, verify= False)
code_site.text
```

```python
soup = BeautifulSoup(code_site.text, 'lxml')
```

```python
soup.find('table',class_='table table-bordered table-hover')
```

```python
def take_manager_iin (link):
    manager_site = session.get(link, verify=False)
    manager_soap = BeautifulSoup(manager_site.text,'lxml')
    manager_iin = manager_soap.find('div',class_='page-inner').findAll('div',class_='panel panel-default')[2].find('div',class_='panel-body').find('table',class_='table table-striped').findAll('tr')[0].find('td').text
    manager_name = manager_soap.find('div',class_='page-inner').findAll('div',class_='panel panel-default')[2].find('div',class_='panel-body').find('table',class_='table table-striped').findAll('tr')[2].find('td').text
    manager_info = [manager_name,manager_iin]
    return manager_info
```

```python
list_size = soup.find('table',class_='table table-bordered table-hover').find('tbody').findAll('tr')
len(list_size)
```

```python
data = []
```

```python
for i in range(len(list_size)):
    link = soup.find('table', class_='table table-bordered table-hover').find('tbody').findAll('tr')[i].findAll('td')[1].find('a').get('href')
    name = soup.find('table', class_='table table-bordered table-hover').find('tbody').findAll('tr')[i].findAll('td')[1].find('strong').text
    bin = soup.find('table', class_='table table-bordered table-hover').find('tbody').findAll('tr')[i].findAll('td')[2].text
    information = take_manager_iin(link)
    manager_name = information[0]
    manager_iin = information[1]
    
    data.append([name,bin,manager_name,manager_iin,link])
```

```python
data
```

```python
import pandas as pd
```

```python
header = ['Наименование организации','БИН организации','ФИО руководителья','ИИН руководителя','Полный адресс организации']
df = pd.DataFrame(data, columns =header)
df_clear = df.drop_duplicates(subset=['Наименование организации'])
df.to_excel('all_data.xlsx',encoding='utf8',index=False)
```
